```python
#import libraries

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
#read file and create data frame
df = pd.read_csv('../../data/internet_usage.csv')
```

```python
#Sanity Checks
print(df.size)
print(df.shape)
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
```

```python
df.rename(columns = {'individuals using the Internet (% of population)': 'internet_usage'}, inplace = True)
```

```python
df_2016=df.query('Year == 2016')
```

```python
df_2016.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df_2016.describe()
```

```python
#create chopleth in Plotly
df_2016.rename(columns = {'Individuals using the Internet (% of population)': 'internet_usage'}, inplace = True)
fig = px.choropleth(df_2016,
                    locations='Code',
                    color='internet_usage',
                    hover_name='Country',
                    color_continuous_scale=px.colors.sequential.Jet,
                    width = 1000,
                    height = 500)
fig.show()
```

```python
#create chloropleth in Plotly
fig = px.choropleth(df_2016,
                    locations='Code',
                    color='internet_usage',
                    hover_name='Country',
                    color_continuous_scale=px.colors.sequential.Jet,
)

fig.update_layout (
    title_text = 'Internet Usage in 2016',
    width = 1000,
    height = 1000,
    geo_scope = 'asia'
)

fig.show()
```

```python
# explore projections in plotly
fig = px.choropleth(df_2016,
                    locations='Code',
                    color='internet_usage',
                    hover_name='Country',
                    color_continuous_scale=px.colors.sequential.Jet,
)

fig.update_layout (
    title_text = 'Internet Usage in 2016',
    width = 1000,
    height = 1000,
    geo={'projection': {'type': 'hammer'}}
)

fig.show()
```

```python
#add animation in plotly

df.sort_values(by = ['Year'], inplace = True)
df.head()
```

```python
df.tail(5)
```

```python
#add animations

# Rename the column in the full DataFrame for consistency
df.rename(columns={'Individuals using the Internet (% of population)': 'internet_usage'}, inplace=True)

fig = px.choropleth(df,
                    locations='Code',
                    color='internet_usage',
                    hover_name='Country',
                    animation_frame = 'Year',
                    color_continuous_scale=px.colors.sequential.Jet,
                    width = 1000,
                    height = 1000
)

fig.update_layout (
    title_text = 'Internet Usage Over Time',
    width = 1000,
    height = 1000,
    geo={'projection': {'type': 'hammer'}}
)

fig.show()
```
