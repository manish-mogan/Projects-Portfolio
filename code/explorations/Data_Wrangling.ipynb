```python
#find the missing data
missing_data = df.isnull()
```

```python
# count missing values in each column
for column in missing_data.columns.values:
  print(column)
  print(missing_data[column].value_counts())
  print(' ')
```

```python
# compute mean and median for normalized-losses
avg_norm_loss = df['normalized-losses'].astype('float').mean(axis = 0)
median_norm_loss = df['normalized-losses'].astype('float').median(axis = 0)


print('Average of normalized-losses:', avg_norm_loss)
print('Median of normalized-losses:', median_norm_loss)
```

```python
# replace missing normalzied - losses with mean

df['normalized-losses'] = df['normalized-losses'].replace(np.nan, avg_norm_loss)
```

```python
#check replacement

df.info()
```

```python
# find the frequency of values for num-of-doors
df['num-of-doors'].value_counts()
```

```python
#find the mode of the distribution for number of doors
df['num-of-doors'].mode()

#or

df['num-of-doors'].value_counts().idxmax()
```

```python
#replace the missing values for number of doors with the mode of the distribution
df['num-of-doors'] = df['num-of-doors'].replace(np.nan, 'four')


df.loc[df['num-of-doors'] == np.nan, 'num-of-doors'] = 'four'
```
