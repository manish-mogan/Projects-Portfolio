```python
#create a series from a list
a = [3,1,4]
p = pd.Series(a)

print(a)

#create a series from a dictionary
b = {"A": 80, "B": 30, "C": 16, "D": 8, "F":12}

q = pd.Series(b)
print(q)
```

```python
#accessing elements in a Series
print(p[1])
print(q["A"])


#updating values in a Series
p[1] = 7
q["B"] = 32
print(p)
print(q)
```

Data Frames

```python
gdp = {
    "USA": [21.5, 21.3, 23.6, 25.7, 27.4],
    "China": [14.3, 14.7, 17.8, 17.9, 17.8],
    "Germany": [3.9, 3.9, 4.3, 4.1, 4.5],
    "Japan": [5.1, 5.1, 5.0, 4.3, 4.2],
    "India": [2.8, 2.7, 3.2, 3.3, 3.5]
}

df = pd.DataFrame(gdp, index = [2019, 2020, 2021, 2022, 2023])
print(df)
print()

df2 = df.transpose()
print(df2)
```

```python
#get index of the dataframes
print(df.index)
print(df2.index)


# get the columns of the dataframe
print(df.columns)
print(df2.columns)
```

```python
#get items by index
print(df.loc[2020])
print(df2.loc['Japan'])
```

```python
#get items by location
print(df.iloc[2])
print()
print(df2.iloc[1])
```
