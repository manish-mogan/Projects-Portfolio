# Datasets

This folder contains a small set of general-purpose datasets used for visualization and EDA examples.

Data files live in `data/`.


```python
from pathlib import Path

import pandas as pd

data_dir = Path("data")
rows = []
for p in sorted(data_dir.iterdir()):
    if p.is_file():
        rows.append({
            "file": p.name,
            "bytes": p.stat().st_size,
            "ext": p.suffix.lower(),
        })

df = pd.DataFrame(rows).sort_values("file").reset_index(drop=True)
df
```
