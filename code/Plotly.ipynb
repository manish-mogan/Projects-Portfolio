```python
import numpy as np
import plotly.graph_objects as go
import plotly.express as px


# generate random numbers
import plotly.io as pio
import plotly.offline as offline


np.random.seed(1)

x = np.random.rand(100)
y = np.random.rand(100)

fig = go.FigureWidget([go.Scatter(x=x, y=y, mode='markers')])

scatter = fig.data[0]
colors = ['#a3a7e4'] * 100
scatter.marker.color = colors
scatter.marker.size = [10] * 100
fig.layout.hovermode = 'closest'

#create our callback function
def update_point (trace, points, selector):
    c = list(scatter.marker.color)
    s = list(scatter.marker.size)
    for i in points.point_inds:
        c[i] = '#bae2be'
        s[i] = 20
        with fig.batch_update():
            scatter.marker.color = c
            scatter.marker.size = s

scatter.on_click(update_point)
fig.show()
```

```python
import numpy as np
import pandas as pd

np.random.seed(42)

data = {
    'x': [1, 2, 3, 4, 5],
    'y': [10, 11, 12, 13, 14],
    'z': [20, 21, 22, 23, 24]
}

df = pd.DataFrame(data)
```

```python
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=2, subplot_titles=("X vs Y", "X vs Z"))
```

```python
fig.add_trace(go.Scattergl(x=df['x'], y=df['y'], mode='markers', name='X vs Y'),
              row=1, col=1)

fig.add_trace(go.Scattergl(x=df['x'], y=df['z'], mode='markers', name='X vs Z'),
              row=1, col=2)
```

```python
def update_selection(trace, points, selector):
    # Get the indices of the selected points
    selected_indices = points.point_inds

    # Determine which trace was selected (0 for X vs Y, 1 for X vs Z)
    selected_trace_index = fig.data.index(trace)
    other_trace_index = 1 - selected_trace_index # The other trace

    # Reset marker properties for all points in the other trace
    colors = ['#a3a7e4'] * len(df)
    sizes = [10] * len(df)

    # Update marker properties for selected points in the other trace
    for i in selected_indices:
        colors[i] = '#bae2be'
        sizes[i] = 20

    # Update the other trace's marker properties in a batch update
    with fig.batch_update():
        fig.data[other_trace_index].marker.color = colors
        fig.data[other_trace_index].marker.size = sizes

# Attach the callback function to the 'on_selection' event for both scatter plots
fig.data[0].on_selection(update_selection)
fig.data[1].on_selection(update_selection)

fig.update_layout(dragmode='select')
fig.show()
```
