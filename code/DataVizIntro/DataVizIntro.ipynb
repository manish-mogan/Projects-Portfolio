##### 3D Plots

```python
#draws a 3d spiral

from mpl_toolkits.mplot3d import Axes3D

# Create a figure and a 3D axes object
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

#generate data for a 3d spiral
t = np.linspace(0, 4 * np.pi, 100)
x = np.sin(t)
y = np.cos(t)
z = t

ax.plot(x, y, z)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

#set a title
ax.set_title('3D Spiral')

#display
plt.show()
```

```python
# create a surface in 3D

#create a meshgrid of X and Y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# define the function to plot (e.g. a saddle shape)
Z = X**2 - Y**2

#create a 3D surface plot
fig = plt.figure(figsize = (10, 8))
ax = fig.add_subplot(111, projection = '3d')

ax.plot_surface(X, Y, Z, cmap = 'viridis')

#add title
ax.set_title('3D Surface Plot')

plt.show()
```

##### Multiple Plots in a Single Figure

```python
# create the data
x = np.linspace(0, 2* np.pi, 100)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

#create a 2x2 grid of subplots
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (10,8))

#plot the first subplot - top_left - axes[0,0]
axes[0,0].plot(x, y1, color = 'blue')
axes[0,0].set_title('Sine Function')

#plot the second subplot - top_right - axes[0,1]
axes[0,1].plot(x, y2, color = 'green')
axes[0,1].set_title('Cosine Function')

#plot the third subplot - bottom_left - axes[1,0]
axes[1,0].plot(x, y3, color = 'red')
axes[1,0].set_title('Tangeant Function')


#plot the fourth subplot - bottom right - axes[1,1]
axes[1,1].plot(x, y4, color = 'blue')
axes[1,1].set_title('Exponential Function')

plt.tight_layout()
fig.suptitle('Math functions')
plt.show()
```

###### Animations with MatPlotLib

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import display, Javascript, HTML


fig, ax = plt.subplots()
x_data, y_data = [], []
line, = ax.plot([], [], 'r-') # Initialize line with empty lists

def animate(i):
  x_data.append (i)
  y_data.append (i**2)
  line.set_data(x_data, y_data)
  ax.set_xlim(0, i+1)
  ax.set_ylim(0, (i+1)**2)
  return line,


#assign the animation to persistent variable 'anim'
anim = FuncAnimation (
    fig,
    animate,
    frames = 200,
    interval = 20,
    blit = True
)


# close the fig
plt.close(fig)


# display as JavaScript
HTML (anim.to_jshtml())
```

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import display, Javascript, HTML


#define time interval
t = np.linspace(0, 10, 100)

#define a sine wave
y = np.sin(t)


fig, axis = plt.subplots()
axis.set_xlim(min(t), max(t))
axis.set_ylim(-2, 2)


animated_plot, = axis.plot([], [])

def update_frame(frame):
  animated_plot.set_data(t[:frame], y[:frame])
  return animated_plot,

anim = FuncAnimation(
    fig = fig,
    func = update_frame,
    frames = len(t),
    interval = 25,
    blit = True
)


#close the figure
plt.close(fig)

HTML(anim.to_jshtml())
```

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import display, Javascript, HTML

# physical constants
m = 1
k = 1
d = 0.2

#solution to the damped oscillator
t = np.linspace(0, 40, 500)
w_d = np.sqrt((4 * m * k - d**2)) / (4 * m ** 2)
x = np.exp(-d * t / (4 * m)) * np.cos(w_d * t)

fig, ax = plt.subplots()
ax.set_xlim(min(t), max(t))
ax.set_ylim(min(x), max(x))

animated_plot, = ax.plot([], [], color='r')

def update_frame(frame):
    animated_plot.set_data(t[:frame], x[:frame])
    return animated_plot,

anim = FuncAnimation(
    fig = fig,
    func = update_frame,
    frames = len(t),
    interval = 25,
    blit = True
)

#close the figure
plt.close(fig)

HTML(anim.to_jshtml())
```
