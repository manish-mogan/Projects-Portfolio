#####Array Creation

```python
#create a 1-D array
a = np.array([2, 8, 3])
print(a)

#create a 2-D array
b = np.array([[4, 9, 2], [3, 5, 7], [8, 1, 6]])
print(b)

#check dimensions
print(a.ndim)
print(b.ndim)

# create an array with all zeros
c = np.zeros(3, dtype = int)
print(c)

# create an array with all ones
d = np.ones(3, dtype = int)
print(d)

# create an array with regularly spaced numbers
e = np.linspace(1, 10, 10)
print(e)


# create a full array with user defined values
f = np.full((2,3), 9, dtype = int)
print(f)


#creaete an array with a range of integers
g = np.arange(10) + 23
print(g)
```

```python
#single graph
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import display, Javascript, HTML


#define time interval
t = np.linspace(0, 10, 100)

#define a sine wave
y = np.sin(t)


fig, axis = plt.subplots()
axis.set_xlim(min(t), max(t))
axis.set_ylim(-2, 2)


animated_plot, = axis.plot([], [])

def update_frame(frame):
  animated_plot.set_data(t[:frame], y[:frame])
  return animated_plot,

anim = FuncAnimation(
    fig = fig,
    func = update_frame,
    frames = len(t),
    interval = 25,
    blit = True
)


#close the figure
plt.close(fig)

HTML(anim.to_jshtml())
```

```python
#double graph
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import display, Javascript, HTML


#define time interval
t = np.linspace(0, 10, 100)

#define a sine wave
y = np.sin(t)
y2 = np.cos(t)

fig, axis = plt.subplots()
axis.set_xlim(min(t), max(t))
axis.set_ylim(-2, 2)


#change this to add - add one for sin and one for cos with ,label = 'sin' or 'cos'
animated_plot_sin, = axis.plot([], [])
animated_plot_cos, = axis.plot([], [])


def update_frame(frame):
  #add animated_plot for sin and cosine. instead of y, write y1 and y2 for sin and cos functions
  animated_plot_sin.set_data(t[:frame], y2[:frame])
  animated_plot_cos.set_data(t[:frame], y[:frame])
  # return both
  return animated_plot_sin, animated_plot_cos,

anim = FuncAnimation(
    fig = fig,
    func = update_frame,
    frames = len(t),
    interval = 25,
    blit = True
)


#close the figure
plt.close(fig)

HTML(anim.to_jshtml())
```

```python
#manual static graph

# import libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import display, Javascript, HTML

# physical constants
m = 1
k = 1
d = 0.2

#solution to the damped oscillator
t = np.linspace(0, 40, 500)
w_d = np.sqrt((4 * m * k - d**2)) / (4 * m * 2)
x = np.exp(-d * t / (4 * m)) * np.cos(w_d * t)

plt.plot(t, x, color = 'r')
plt.show()
```

```python
#animated dynamic graph

# import libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import display, Javascript, HTML

# physical constants
m = 1
k = 1
d = 0.2

#solution to the damped oscillator
t = np.linspace(0, 40, 500)
w_d = np.sqrt((4 * m * k - d**2)) / (4 * m * 2)
x = np.exp(-d * t / (4 * m)) * np.cos(w_d * t)


#animation starting here **

#write fig and x and plt.subplots
fig, ax = plt.subplots()
ax.set_xlim(min(t), max(t))
ax.set_ylim(min(x), max(x))


#create the animated plot
animated_plot, = ax.plot([], [], color='r')


#define the update frame function
def update_frame(frame):
    animated_plot.set_data(t[:frame], x[:frame])
    return animated_plot,


anim = FuncAnimation(
    fig = fig,
    func = update_frame,
    frames = len(t),
    interval = 25,
    blit = True
)





plt.plot(t, x, color = 'r')
plt.show()
```
